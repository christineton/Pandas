```python
# life expectation less than 30. single column, then rows in this. Boolean subsetting
df.loc[df['lifeExp'] < 30,['country','year','pop']]
```

```python
# life expectation less than 30, population less than 7 million
df.loc[(df['lifeExp'] < 30) & (df['pop'] < 7000000), ['country','year','pop']]
```

```python
# life expectation less than 30, population less than 8 million
df.loc[(df['lifeExp'] < 30) & (df['pop'] < 8_000_000), ['country','year','pop']]
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# filter rows by smoker = 'no' and total bill >= 10
tips.loc[(tips['smoker'] == 'No') & (tips.total_bill >= 10)]
```

```python
# what is the average total_bill for each value of smoker, day and time
col = tips.groupby(['smoker', 'day', 'time'])(['total_bill']).mean()
```
